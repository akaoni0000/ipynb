```python
from rapidfuzz import fuzz
import pandas as pd

# 入力データ（例として一部だけ記載）
projects = [
    "AWSプロジェクト 10月期",
    "AWS移行プロジェクト 冬期",
    "AWS最適化PJ 2025",
    "AWSクラウド推進PJ 第2弾",
    "AWSセキュリティ強化プロジェクト",
    "Azure導入プロジェクト 春期",
    "Azure最適化PJ 2025",
    "Azure移行支援プロジェクト",
    "DX推進プロジェクト 第1弾",
    "DX推進プロジェクト 秋期",
    "DX改革プロジェクトA",
    "システム統合PJ",
    "基幹システム刷新PJ",
]

# 類似度しきい値
threshold = 70

# 結果を格納
groups = []
used = set()

for i, p1 in enumerate(projects):
    if i in used:
        continue
    group = [p1]
    used.add(i)
    for j, p2 in enumerate(projects):
        if j in used:
            continue
        if fuzz.partial_ratio(p1, p2) > threshold:
            group.append(p2)
            used.add(j)
    groups.append(group)

# グループの代表名（最初のものを採用）
mapping = {}
for group in groups:
    new_name = group[0]
    for old in group:
        mapping[old] = new_name

# 出力（CSVのように表示）
df = pd.DataFrame(list(mapping.items()), columns=["古いプロジェクト名", "新しいプロジェクト名"])
print(df)
```

```python
import MeCab

def tokenize(text):
    tagger = MeCab.Tagger("-Owakati")  # -Owakati は分かち書きモード
    return tagger.parse(text).strip().split()

def simple_similarity(text1, text2):
    words1 = set(tokenize(text1))
    words2 = set(tokenize(text2))
    
    common = words1 & words2  # 共通単語
    all_words = words1 | words2  # 全部の単語
    
    if len(all_words) == 0:
        return 0.0
    
    similarity = len(common) / len(all_words)
    return similarity


text1 = "私は犬が好きです"
text2 = "私は猫が好きです"
text3 = "今日は天気がいいです"

print("1 vs 2:", simple_similarity(text1, text2))
print("1 vs 3:", simple_similarity(text1, text3))

```

```python
import MeCab
import csv

# -----------------------------
# ① MeCabで単語を分割
# -----------------------------
def tokenize(text):
    tagger = MeCab.Tagger("-Owakati")
    return set(tagger.parse(text).strip().split())

# -----------------------------
# ② 類似度を共通単語割合で計算
# -----------------------------
def similarity(a, b):
    wa, wb = tokenize(a), tokenize(b)
    common = wa & wb
    total = wa | wb
    return len(common) / len(total) if total else 0

# -----------------------------
# ③ 類似したプロジェクトをグループ化
# -----------------------------
def group_similar_projects(projects, threshold=0.5):
    groups = []
    for p in projects:
        found = False
        for group in groups:
            if similarity(p, group[0]) >= threshold:
                group.append(p)
                found = True
                break
        if not found:
            groups.append([p])
    return groups

# -----------------------------
# ④ プロジェクトリスト
# -----------------------------
projects = [
    "ディスク返却不要サービス 警視庁 5月",
    "ディスク返却不要サービス 警視庁 6月",
    "ディスク返却不要サービス 警視庁 7月",
    "官公庁 2022年4月~8月 マルチベンダHW統括サービス",
    "官公庁 2022年9月~12月 マルチベンダHW統括サービス",
    "官公庁 2024年3月~10月 マルチベンダHW統括サービス",
    "官公庁 2025年2月~10月 マルチベンダHW統括サービス",
    "UPSバッテリ交換サービス IHI様 30台 2022年4月",
    "UPSバッテリ交換サービス IHI様 10台 追加 2022年12月"
]

# -----------------------------
# ⑤ グルーピング実行
# -----------------------------
groups = group_similar_projects(projects, threshold=0.5)

# -----------------------------
# ⑥ CSV出力
# -----------------------------
with open("project_groups.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    for group in groups:
        representative = group[0]
        for p in group:
            writer.writerow([p, representative])

print("✅ 出力完了: project_groups.csv")
```
